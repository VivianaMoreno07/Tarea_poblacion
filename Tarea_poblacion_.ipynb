```python
import pandas as pd
poblacion=pd.read_csv("osb_demografia-poblacion-localidad.csv",encoding="latin1",delimiter=";")
poblacion.head()
```

```python
#obtener la poblacion para el 2024
poblacion_2024= poblacion[poblacion['Año']==2024]
poblacion_2024.head()
```

```python
#agrupar los datos por localidad
grupos=poblacion_2024.groupby("Localidad")
print(grupos)
#sumar cada localidad
total_poblacion_localidad=grupos["Población"].sum()
print(total_poblacion_localidad)
```

```python
#se crea una serie llamada Población con los nombre4s de las localidades como indice y la su
#como valores

#convertir la serie a dataframe.El modo reset_index () aplicado a la serie, cambie el
#Convertir la serie a dataframe
df_poblacion_localidad=total_poblacion_localidad.reset_index()
df_poblacion_localidad
```

```python
df_poblacion_localidad["Localidad"]=df_poblacion_localidad["Localidad"].str.upper()
df_poblacion_localidad
```

```python
def replace_accents(input_str):
    replacements={
        'Á':'A',
        'É':'E',
        'Í':'I',
        'Ó':'O',
        'Ú':'U',
    }
    for accented_char,unaccented_char in replacements.items():
        input_str=input_str.replace(accented_char,unaccented_char)
    return input_str

df_poblacion_localidad["Localidad"]=df_poblacion_localidad["Localidad"].apply(replace_accents)
df_poblacion_localidad
```

```python
# Suponiendo que 'Centroides' es el nombre de una nueva columna que deseas agregar al DataFrame 'gdf_localidades'
# Primero, calcula los centroides de cada polígono en 'gdf_localidades'
gdf_localidades['Centroides'] = gdf_localidades.centroid

# Luego, plotea el mapa y los centroides
mapa = gdf_localidades.plot(figsize=(10,10))
gdf_localidades["Centroides"].plot(ax=mapa, color="white")
```

```python
import contextily as cx

mapa_2=gdf_localidades.plot(
    column="LocNombre",
    figsize=(15,15),
    edgecolor="black",
    cmap="tab20",
    legend=True,
    legend_kwds={
        "loc":"upper left"
    }

)


cx.add_basemap(mapa_2,crs=gdf_localidades.crs.to_string())
```

```python
# Suponiendo que las columnas que deseas utilizar para unir los DataFrames se llaman 'codigo_localidad' en gdf_localidades y 'codigo_localidad' en poblacion
merged_df =gdf_localidades.merge(df_poblacion_localidad,left_on='LocNombre', right_on='Localidad')
merged_df.drop('Localidad',axis=1,inplace=True)
```

```python
# Verifica si merged_df ya es un GeoDataFrame, si no, conviértelo
if not isinstance(merged_df, gpd.GeoDataFrame):
    merged_df = gpd.GeoDataFrame(merged_df, geometry='geometry')
```

```python
import contextily as cx
import matplotlib.pyplot as plt

# Configurar el tamaño del gráfico
fig, ax = plt.subplots(figsize=(10, 10))  # Cambiar el tamaño del gráfico

# Graficar el GeoDataFrame, coloreando por la columna 'Población' y ajustando el color del mapa
merged_df.plot(column='Población', cmap='YlGnBu', legend=True, ax=ax)  # Cambiar el color del mapa

# Agregar el mapa base con un tamaño de zoom diferente
cx.add_basemap(ax, crs=merged_df.crs.to_string(), source=cx.providers.OpenStreetMap.Mapnik, zoom=10)  # Cambiar el tamaño del mapa

# Agregar etiquetas de población en el centroide de cada polígono
for idx, row in merged_df.iterrows():
    centroid = row['geometry'].centroid
    ax.text(centroid.x, centroid.y, row['Población'], fontsize=6, ha='center', va='center', color='black')  # Cambiar el tamaño del texto
# Agregar leyenda
plt.legend(title='Localidades', loc='lower right')
#
# Agregar título y mostrar el gráfico
plt.title('Mapa de Localidades con Población')
plt.show()


```
